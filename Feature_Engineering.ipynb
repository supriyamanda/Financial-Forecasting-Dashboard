```python
import pandas as pd
df=pd.read_csv('/content/all_stocks_5yr.csv')
df.head()
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df['Daily Return']=df['close'].pct_change()
```

```python
df['7-Day MA'] = df['close'].rolling(window=7).mean()
df['30-Day MA'] = df['close'].rolling(window=30).mean()
```

```python
df['Cumulative Return'] = (1 + df['Daily Return']).cumprod() - 1
```

```python
df['Volatility'] = df['Daily Return'].rolling(window=30).std()
```

```python
df=df.dropna()
```

```python
df.reset_index(inplace=True)
```

```python
df.to_csv('/content/all_stocks_5yr_feature_engineered.csv', index=False)
```
